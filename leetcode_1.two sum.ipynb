```python
nums = [2,7,11,15]
target = 9
```

```python
# my solution
class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        for i in range(len(nums)):
            for j in range(i+1,len(nums)):
                sum_num = nums[i]+nums[j]
                if sum_num == target:
                    return [i,j]        
```

```python
nums = [3,3]
target = 6
```

```python
# use dictionary
dic = {}
for index, val in enumerate(nums):
    dic[val] = index
    
for idx,val in enumerate(nums):
    minus = target-val
    if minus in dic and (idx != dic[minus]):
        print(idx,dic[minus])
        break
    
```
